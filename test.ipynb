Import data

```python
gd1 = pd.read_csv('./data/Plant_1_Generation_Data.csv')
wd1 = pd.read_csv('./data/Plant_1_Weather_Sensor_Data.csv')
gd2 = pd.read_csv('./data/Plant_2_Generation_Data.csv')
wd2 = pd.read_csv('./data/Plant_2_Weather_Sensor_Data.csv')
```

Convert DATE_TIME column

```python
gd1['DATE_TIME'] = pd.to_datetime(gd1["DATE_TIME"])
wd1['DATE_TIME'] = pd.to_datetime(wd1["DATE_TIME"])
gd2['DATE_TIME'] = pd.to_datetime(gd2["DATE_TIME"])
wd2['DATE_TIME'] = pd.to_datetime(wd2["DATE_TIME"])
```

```python
# information from data
plant1_inverters = gd1['SOURCE_KEY'].nunique()
plant2_inverters = gd2['SOURCE_KEY'].nunique()
print('Plant 1 has {} inverters\nPlant 2 has {} inverters'.format(plant1_inverters, plant2_inverters))

axes = sns.barplot(data=gd1, x='SOURCE_KEY', y='AC_POWER', orient='v', color='blue', ci=None)
axes.set_title('Plant 1 Mean AC Power Output per Inverter')
axes.set(ylabel='Mean AC Power Output (kW)')
axes.set_xticklabels(labels=gd1['SOURCE_KEY'].unique(), rotation=90)
plt.show()

axes = sns.barplot(data=gd2, x='SOURCE_KEY', y='AC_POWER', orient='v', color='blue', ci=None)
axes.set_title('Plant 2 Mean AC Power Output per Inverter')
axes.set(ylabel='Mean AC Power Output (kW)')
axes.set_xticklabels(labels=gd2['SOURCE_KEY'].unique(), rotation=90)
plt.show()
```

```python
# merge data
plant1 = gd1.merge(wd1, right_on='DATE_TIME', left_on='DATE_TIME')
plant1 = plant1.drop(['PLANT_ID_y', 'SOURCE_KEY_y'], axis=1)
plant1['TIME']=plant1['DATE_TIME'].dt.time
plant1['DATE']=plant1['DATE_TIME'].dt.date

plant2 = gd2.merge(wd2, right_on='DATE_TIME', left_on='DATE_TIME')
plant2 = plant2.drop(['PLANT_ID_y', 'SOURCE_KEY_y'], axis=1)
plant2['TIME']=plant2['DATE_TIME'].dt.time
plant2['DATE']=plant2['DATE_TIME'].dt.date

print(plant1.info())
print(plant2.info())
```
